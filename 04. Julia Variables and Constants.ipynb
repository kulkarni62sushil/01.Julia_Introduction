# 4.1. Simple Program

Start with a simple program to calculate the area and the perimeter of the rectangle given the length and breadth.

```julia
l= 10 
b=20 

area= l* b
perimeter= 2*(l + b)
print("Area =",area) 
print("  Perimeter =",perimeter )
```

Let us understand the code line by line:

* The first statement tells that the **variables** *l*, *b*, *area*, *perimeter* are declared as integer. 

* The next statement assigns the **constants** 10 and 20 to *l* and *b* and stores these values.

* Next the statements `area = l*b` and `perimeter= 2*(l + b)` assigns the value of the right hand side expression to the variable *`area`* and *`perimeter`* i.e. `l* b` i.e.`10 * 20` viz. 200 and 
`2*(l + b)` i.e. 2*(10 + 20) viz. 60. Hence *`area`* is 200 and *`perimeter`* is 60 of the rectangle. 

* The function **`print( )`** displays the result on the screen.

With reference to the above simple example let us understand the standard words used above in bold letters in detail.

## 4.2.2.	Identifiers

In the above program the elements *`l`*, *`b`*, *`area`*, *`perimeter`* and *`print`*( ) are called **identifiers.**

Thus identifiers are names given to various program elements, such as variables, classes, functions,arrays etc. Identifier consists of series of letters, digits and underscore in any order except that the first character must be a letter or an underscore (_).

## 4.2.3. Constants

In the above program 10 and 20 are called **constants**.

A constant or literal is an entity whose value does not change during program execution. For example, 11, 23.45, -532, ‘a’, ‘T’, “Aishwarya” are few constants. Constants may be numeric, character, string or boolean.

**Numeric constants** consist of positive integers, negative integers, zero and fractions. Thus 11, 23.45 are numeric constants. There are two types of numeric constants: integer constants and floating point constants.

**Integer constants** are constants which are positive integers, negative integers and zero. On the other hand, Floating point constants are constants which are fractions and numbers with decimal points.

**Character constants** are constants that take just a single character and written in a single quotation. For example, ‘e’ or ‘10’ etc.

**String constants** are the list of characters and are written in double quotation. For example, “Aishwarya”, “Swati is brilliant”, “ Payal got 99 marks out of 100 “ etc.

**Boolean constants** are constants that hold only truth values true or false.

## 4.2.4.	Variables



In the above program `*l*`, `*b*`, `*area*` and `*perimeter*` are called **variables**.

**Variables** are entities used by a program to store values in the computation. Variable names are names given to the location in the memory where different constants are stored. The locations can contain constants like integer, float, character or boolean. The type of variables depends on the types of constants that it stores. For example, an integer variable stores integer constant, a character variable stores a character constant and so on.

Julia places some restriction on what can be a variable name. Following are the rules for constructing variable names, which are same as the rules for identifiers:

A variable name must begin with a letter, dollar symbol (‘$’) or an underscore (‘-’), which may be followed by a sequence of letters or digits (0-9),‘$’,‘-’. No commas or blanks are allowed within a variable name.

Thus one can write variables as *`$length`* or *`_length`* or *`length2`* or *`length_2`* etc. 

**Following are conventions to write variable names in Julia:**

- A variable name should not be a Julia key word. For example you cannot declare a variable called *`print`* or*`length*`.

- Variable name must be unique. For example, to store three different numbers, three unique variable names need to be used.
 
- The variable name must be meaningful and short .The name must reflect the data that the variable contains. For example, to store age of an employee, the variable name could be just *`age`* or *`employeeAge`*.

- Variable names must be nouns and possibly starts with a lowercase letter.

- If variable name contains two or more words, join these words and begin each word with an uppercase letter. The first word, however, starts with a lowercase letter. For example, to store marks in mathematics one can take the variable name as *`marksMath`*.

- As Julia is case sensitive so *`employeeSalary`* and *`EmployeeSalary`* are two different variable names.

**Following are few valid variable names:**

(i) *`age`* , (ii) *`marks`*, (iii) *`myMarks`*, (iv) *`incomeTax`*, (v) *`thisVariableNameContainsMyName`*.
(vi)*`$age`*, (vii) *`aim$`*, (viii) *`_total`*.

Following are few invalid variable names:

(i) *`%name`*, (ii) `*marks` , (iii) *`Employee#name`*
